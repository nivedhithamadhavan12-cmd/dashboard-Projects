```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("HR_PJ_messy_dataset.csv")
df
```

```python
print(df.head())
print(df.shape)
print(df.info())
```

```python
df_original = df.copy()
```

```python
df.columns=(df.columns.str.strip().str.replace(" ","_"))
df.rename(columns={"Unnamed:_0":"EmployeeID"},inplace=True)
df
```

```python
duplicate=df.duplicated().sum()
print("Dupplicate Rows",duplicate)
```

```python
df.isnull().sum()
```

```python
df["Age"]=pd.to_numeric(df["Age"],errors="coerce")
df["Age"]=df["Age"].fillna(df["Age"].median())
df["Age"]=df["Age"].astype(int)
df.loc[df["Age"]<18,"Age"]=19
df.loc[df["Age"]>60,"Age"]=60
df.isnull().sum()
```

```python
text_columns=["Gender","Department","Job_Title","Education","Attrition","Region","Payment_Method"]
for col in text_columns:
    df[col]=(df[col].astype(str).str.strip().str.title().replace("Nan",pd.NA))
df    
    
```

```python
df.dtypes
```

```python
df["Gender"].value_counts(dropna=False)
#df["Gender"] = df["Gender"].replace("Nan", pd.NA)
df["Gender"]=df["Gender"].replace({"Fmale":"Female","Fm":"Female","M":"Male","F":"Female"})
df["Gender"]=df["Gender"].fillna("Female")
df
df.isnull().sum()
```

```python
df["Department"].value_counts(dropna=False)
df["Department"] = df["Department"].replace({"It": "IT","I.T": "IT","Hr": "HR","H.R": "HR","Engg": "Engineering","Mktg": "Marketing","Ops": "Operations"})
m=df["Department"].mode()[0]
print(m)
df["Department"]=df["Department"].fillna(m)
df
df.isnull().sum()
```

```python
df["Job_Title"].value_counts(dropna=False)
df["Job_Title"]=df["Job_Title"].replace({"Assoc":"Associate","Vp":"VP","V.P":"VP","Snr Analyst":"Senior Analyst","Sr Manager":"Senior Manager","Mgr":"Manager",
                                         "Int":"Intern","Dir":"Director","Consult":"Consultant"})
m=df["Job_Title"].mode()[0]
print(m)
df["Job_Title"]=df["Job_Title"].fillna(m)
df["Job_Title"].value_counts(dropna=False)
df
df.isnull().sum()
```

```python
df["Education"].value_counts(dropna=False)
df["Education"] = df["Education"].replace({
    "Bachelors": "Bachelor",
    "Bachelor": "Bachelor",

    "Masters": "Master",
    "Master": "Master",

    "Phd": "PhD",
    "P.H.D": "PhD",

    "Hs": "High School",
    "Highschool": "High School",

    "Assoc Degree": "Associate",
    "Associate": "Associate"
})
df["Education"]=df["Education"].fillna(df["Education"].mode()[0])
df
df.isnull().sum()
```

```python
df["Hire_Date"].value_counts(dropna=False)
df["Hire_Date"]=df["Hire_Date"].replace(["--","TBD","Unknown"],pd.NA)
df["Hire_Date"]=pd.to_datetime(df["Hire_Date"],errors="coerce",dayfirst=True)
df["Hire_Date"]=df["Hire_Date"].fillna(df["Hire_Date"].mode()[0])
df.dtypes
df.isnull().sum()
df
```

```python
print(df["Salary"].value_counts(dropna=False))
df["Salary"] = (
    df["Salary"]
    .astype(str)
    .str.replace(",", "")
    .str.replace("$", "", regex=False))
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
df["Salary"]=df["Salary"].fillna(df["Salary"].mode()[0])
df.isnull().sum()
```

```python
df["Years_At_Company"].value_counts(dropna=False)
df["Years_At_Company"]=df["Years_At_Company"].fillna(0)
df.isnull().sum()
```

```python
df["Job_Satisfaction"]=df["Job_Satisfaction"].fillna(df["Job_Satisfaction"].mode()[0])
df["Job_Satisfaction"].value_counts(dropna=False)
df.isnull().sum()
```

```python
df["Performance_Rating"]=df["Performance_Rating"].fillna(df["Performance_Rating"].mode()[0])
df["Performance_Rating"].value_counts(dropna=False)
#df.isnull().sum()
```

```python
df["Monthly_Hours"].value_counts(dropna=False)
df["Monthly_Hours"] = df["Monthly_Hours"].fillna(df["Monthly_Hours"].median())
df.isnull().sum()
```

```python
df["Attrition"].value_counts()
df["Attrition"] = df["Attrition"].replace({"Left": "Yes","Y": "Yes","Stayed": "No","N": "No"})
df["Attrition"] = df["Attrition"].fillna("No")
df.isnull().sum()
```

```python
df["Region"].value_counts()
mapping = {"Apac": "Asia Pacific","N. America": "North America","Latam": "Latin America","Eur": "Europe","Mid East": "Middle East","Me": "Middle East"}
df["Region"] = df["Region"].replace(mapping)
df["Region"] = df["Region"].fillna(df["Region"].mode()[0])
df.isnull().sum()
```

```python
df["Payment_Method"].value_counts(dropna=False)
mapping = {"Wire Transfer": "Wire Transfer","Wire": "Wire Transfer","Check": "Check","Chk": "Check","Direct Deposit": "Direct Deposit","Dd": "Direct Deposit","Cash": "Cash"}
df["Payment_Method"] = df["Payment_Method"].replace(mapping)
p=df["Payment_Method"].mode()[0]
print(p)
df["Payment_Method"] = df["Payment_Method"].fillna(p)
```

```python
df["Phone"].value_counts(dropna=False)
df=df.drop(columns=["Phone"])
```
